```python
# CALL FUNCTION TO GET ERROR 

def error(ground_truth, lower_res):
    file_paths = {
        1025: "output/1025_0.75t/Soln_0000.0000.h5",
        513: "output/513_0.75t/Soln_0000.0000.h5",
        257: "output/257_0.75t/Soln_0000.0000.h5",
        129: "output/129_0.75t/Soln_0000.0000.h5",
        65: "output/65_0.75t/Soln_0000.0000.h5"
    }

    if ground_truth and lower_res in file_paths:
        GT = hdf5_reader(file_paths[ground_truth], "Vz")
        E = hdf5_reader(file_paths[lower_res], "Vz")
    else:
        return f"No resolutions of this type: {ground_truth} or {lower_res}"
    skip_factor = round((groud_truth-1)/(lower_res-1))
    diff = 0
    count = 0 
    
    for i in range(0, E.shape[0]):
        for j in range(0, E.shape[1]):
            gt_value = GT[i * skip_factor, j * skip_factor]
            e_value = E[i, j]
            diff += np.abs(gt_value - e_value)
            count += 1

    mae = diff/count
    print(f"The MAE for the ground truth {ground_truth} and lower resolution: {lower_res} is: {mae}")
```

```python
def hdf5_reader(file_path, dataset_name):
    with h5py.File(file_path, 'r') as f:
        data = f[dataset_name][:]
    return data

def point_wise_error(ground_truth, lower_res):
    file_paths = {
        1025: "output/1025_0.75t/Soln_0000.0000.h5",
        513: "output/513_0.75t/Soln_0000.0000.h5",
        257: "output/257_0.75t/Soln_0000.0000.h5",
        129: "output/129_0.75t/Soln_0000.0000.h5",
        65: "output/65_0.75t/Soln_0000.0000.h5"
    }



    if ground_truth in file_paths and lower_res in file_paths:
        GT = hdf5_reader(file_paths[ground_truth], "Vx")
        E = hdf5_reader(file_paths[lower_res], "Vx")
    else:
        return f"No resolutions of this type: {ground_truth} or {lower_res}"

    
    plt.figure(figsize=(8, 6))

    plt.imshow(GT, cmap='hsv')  
    plt.colorbar(label='Values')  

    plt.title("Ground Truth Vx", fontsize=14, color='midnightblue', fontweight='bold')
    plt.xlabel("X Axis", fontsize=12)
    plt.ylabel("Y Axis", fontsize=12)

    plt.savefig("groundTruth.png", dpi=300)  
    plt.show()
    
    x_small = np.linspace(0, 1, E.shape[0])
    y_small = np.linspace(0, 1, E.shape[1])
    interpolator = RegularGridInterpolator((x_small, y_small), E)

    x_large = np.linspace(0, 1, GT.shape[0])
    y_large = np.linspace(0, 1, GT.shape[1])
    points_large = np.array(np.meshgrid(x_large, y_large)).T.reshape(-1, 2)
    E_interpolated = interpolator(points_large).reshape(GT.shape)

    error = GT - E_interpolated
    plt.figure(figsize=(8, 6))

    plt.imshow(error, cmap='hsv')  
    plt.colorbar(label='Error Magnitude')  

    plt.title("Point-wise Error Vx", fontsize=14, color='midnightblue', fontweight='bold')
    plt.xlabel("X Axis", fontsize=12)
    plt.ylabel("Y Axis", fontsize=12)

    plt.savefig("pointWiseError.png", dpi=300)  
    plt.show()

# Example usage
ground_truth = int(input("Enter the ground truth"))
lower_res = int(input("Enter the lower resolution"))
point_wise_error(ground_truth, lower_res)
```
