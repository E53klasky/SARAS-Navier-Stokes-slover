```python
def hdf5_reader(filename,dataset):
    file_V1_read = h5py.File(filename)
    dataset_V1_read = file_V1_read["/"+dataset]
    V1=dataset_V1_read[:,:]

    return V1
```

```python
def plotStreamline(fName):

    U = hdf5_reader(fName, "Vx")
    #V = hdf5_reader(fName, "Vy")
    W = hdf5_reader(fName, "Vz")
    #T = hdf5_reader(fName, "T")

    [Nx, Nz] = U.shape

    print("Nx",Nx)

    #x = np.arange(0, 10)
    #y = np.arange(0, 10)
    x = np.linspace(0,1,Nx)
    z = np.linspace(0,1,Nz) 

    X, Z = np.meshgrid(x, z)

    print(X.shape)

    plt.figure()

    plt.streamplot(X, Z, U, W, density = 1)
 
    # show plot
    plt.show()
```

```python
def error(ground_truth, lower_res):
    file_paths = {
        1025: "output/1025_0.75t/Soln_0000.7500.h5",
        513: "output/513_0.75t/Soln_0000.7500.h5",
        257: "output/257_0.75t/Soln_0000.7500.h5",
        129: "output/129_0.75t/Soln_0000.7500.h5",
        65: "output/65_0.75t/Soln_0000.7500.h5"
    }

    if ground_truth and lower_res in file_paths:
        GT = hdf5_reader(file_paths[ground_truth], "Vx")
        E = hdf5_reader(file_paths[lower_res], "Vx")
    else:
        return f"No resolutions of this type: {ground_truth} or {lower_res}"
    skip_factor = round((groud_truth-1)/(lower_res-1))
    diff = 0
    count = 0 
    
    for i in range(0, E.shape[0],1):
        for j in range(0, E.shape[1],1):
            gt_value = GT[i * skip_factor, j * skip_factor]
            e_value = E[i, j]
            diff += np.abs(gt_value - e_value)
            count += 1
    mae = diff/count
    print(f"The MAE for the ground truth {ground_truth} and lower resolution: {lower_res} is: {mae}")
```

```python
print("Enter the number of grid points")
try:
    groud_truth = int(input("Enter in the ground truth:"))
    lower_res   = int(input("Enter in the lower resolution you want to compare it to"))
except:
    print("Enter a numeric value!!!")
error(groud_truth, lower_res)
```
